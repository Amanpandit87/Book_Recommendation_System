```python
def recmmend(book_name):
    book_id = np.where(pivot_book.index==book_name)[0][0]
    distance , suggestion = Model.kneighbors(pivot_book.iloc[book_id,:].values.reshape(1,-1),n_neighbors = 6)

    for i in range(len(suggestion)):
        suggested_book_id = suggestion[0][i]
        suggested_book_name = pivot_book.index[suggested_book_id]
        if suggested_book_name == book_name:
            pass
        else:
            print(pivot_book.index[suggestion[i]])
```

```python
recmmend("Animal Farm")
```

```python
def recommend(book_name):
    try:
        # Get the book ID
        book_id = np.where(pivot_book.index == book_name)[0][0]
        
        # Get the neighbors
        distance, suggestion = Model.kneighbors(pivot_book.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)
        
        # Iterate through suggestions
        for i in range(len(suggestion[0])):
            suggested_book_id = suggestion[0][i]
            suggested_book_name = pivot_book.index[suggested_book_id]
            
            # Skip the queried book itself
            if suggested_book_name == book_name:
                continue
            
            print(suggested_book_name)
    except IndexError:
        print(f"Book '{book_name}' not found in the pivot table.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
recommend("Animal Farm")
```
